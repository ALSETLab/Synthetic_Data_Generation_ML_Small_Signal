Traditional Machine Learning Algorithms for Classification
========================

In this notebook, several traditional Machine Learning techniques are trained and tested for the eigenvalue dataset preprocessed in the previous notebooks. The first cell imports the required libraries and the processed training and testing data.

```python
%run -i import_libraries.py
%run -i ml_algorithms.py

with open('99_Data/normalized_testing_training_data_red.pkl', 'rb') as f:
    normalized_testing_training_data_red = pickle.load(f)

X_train_not_red = normalized_testing_training_data_red['X_train']
X_train = normalized_testing_training_data_red['X_train_red']
Y_train = np.reshape(normalized_testing_training_data_red['Y_train_red'], -1)
X_test = normalized_testing_training_data_red['X_test']
Y_test = np.reshape(normalized_testing_training_data_red['Y_test'], -1)
```

# Supervised Learning Algorithms
### Logistic and Softmax Regression

```python
logistic_regression(X_train, X_train_not_red, X_test, Y_train, Y_test)
```

```python
softmax_regression(X_train, X_train_not_red, X_test, Y_train, Y_test)
```

### Support Vector Machines (SVM)

```python
svm(X_train, X_train_not_red, X_test, Y_train, Y_test)
```

### $k$-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier

# Loop to select the 'n_neighbors' hyperparameter
for n in range(1, 6):
    clf = KNeighborsClassifier(n_neighbors = n)
    clf.fit(X_train, Y_train)
    print(f"Number of neighbors: {n}")
    print("Training accuracy: {}".format(clf.score(X_train, Y_train)))
    print("Testing accuracy: {}".format(clf.score(X_test, Y_test)))

kNN(X_train, X_train_not_red, X_test, Y_train, Y_test)
```

### Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz

# Loop for selecting the 'max_depth' hyperparameter
for max_d in range(1,10):
    tree = DecisionTreeClassifier(max_depth = max_d, random_state = 0)
    tree.fit(X_train, Y_train)
    print(f"Depth (max): {max_d}")
    print("Training accuracy: {}".format(tree.score(X_train, Y_train)))
    print("Testing accuracy: {}".format(tree.score(X_test, Y_test)))
    
decision_tree(X_train, X_train_not_red, X_test, Y_train, Y_test)
```

```python
naive_bayes(X_train, X_train_not_red, X_test, Y_train, Y_test)
```
